```python
import chardet
import pandas as pd

```

```python

messages=pd.read_csv('spam.csv',sep=',',encoding='latin1')
```

```python
messages.head()
```

```python
messages.info()
```

```python
#Drop the Columns
messages.drop(columns = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace = True)
```

```python
 # Rename the columns name
messages.rename(columns = {'v1': 'target', 'v2': 'text'}, inplace = True)
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
le=WordNetLemmatizer()
from nltk.corpus import stopwords
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['text'][i])
    review=review.lower()
    review=review.split()
    review=[le.lemmatize(word,pos='v') for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=1000,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
## Create the Bag OF Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=1000,binary=True,ngram_range=(1,3))
X=cv.fit_transform(corpus).toarray()
```
