```python
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Example data
sentences = [
    "The quick brown fox jumps over the lazy dog",
    "I love machine learning and natural language processing"
]
```

```python
# Tokenizing sentences
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]
```

```python
# Training Word2Vec model
model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python

import numpy as np

def average_word_vectors(words, model, num_features):
    # Initialize an empty array for storing the sum of word vectors
    feature_vector = np.zeros((num_features,), dtype="float32")
    num_words = 0
    
    for word in words:
        if word in model.wv:
            num_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])
    
    # Divide the result by the number of words to get the average
    feature_vector = np.divide(feature_vector, num_words)
    return feature_vector

# Compute average word vectors for each sentence
avg_word_vectors = [average_word_vectors(sentence, model, 100) for sentence in tokenized_sentences]
print(avg_word_vectors)
```

```python
long_sentence = "Natural language processing enables computers to understand and process human language, opening up a wide range of applications and opportunities."
```

```python
import numpy as np

def average_word_vectors(words, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    num_words = 0
    
    for word in words:
        if word in model.wv:
            num_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])
    
    if num_words > 0:
        feature_vector = np.divide(feature_vector, num_words)
    return feature_vector

# Tokenize the sentence
tokenized_sentence = word_tokenize(long_sentence.lower())

# Compute the average word vector
avg_vector = average_word_vectors(tokenized_sentence, model, 100)
print(avg_vector)
```
